```python
for i in df.columns:
    if df[i].dtype != 'O':
        plt.figure()
        sns.boxplot(y = i, data = df)
        plt.title('Boxplot of {}'.format(i))
        plt.show()
```

Here only the column 'SW' has outliers.

In all other columns, the values are in the Interqurtile range(IQR).

```python
sns.heatmap(df.corr(numeric_only = True), annot = True, linewidth = 1)
plt.show()
```

```python
for k in df.columns:
    if df[k].dtype != 'O':
        plt.hist(df[k], rwidth = 0.95)
        plt.title('Distribution of {}'.format(k))
        plt.show()
```

```python
df.groupby('Classification').agg(['min', 'max', 'mean'])
```

It can be seen that min,max and mean values in all the columns, except "SW", are highest for the classification of Iris-virginica.
